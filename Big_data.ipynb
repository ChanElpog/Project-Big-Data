```python
# กรองแบรนด์หลักๆ เช่น Apple, Xiaomi, Oppo, Samsung, Huawei
brands = ['Apple', 'Xiaomi', 'Oppo', 'Samsung', 'Huawei']

# สร้าง DataFrame ที่กรองแบรนด์เหล่านี้
df_sales_filtered_brands = df_sales_new[df_sales_new['Manufacturer'].isin(brands)]

# รวมยอดขายตามปีและแบรนด์
df_sales_brands_yearly = df_sales_filtered_brands.groupby(['Year', 'Manufacturer'])['Units Sold (million )'].sum().reset_index()

# สร้างกราฟการเปลี่ยนแปลงยอดขายตามปีสำหรับแต่ละแบรนด์
plt.figure(figsize=(12, 6))

# วาดกราฟสำหรับแต่ละแบรนด์
for brand in brands:
    brand_data = df_sales_brands_yearly[df_sales_brands_yearly['Manufacturer'] == brand]
    plt.plot(brand_data['Year'], brand_data['Units Sold (million )'], marker='o', label=brand)

    # เพิ่มข้อความในแต่ละจุด
    for i in range(len(brand_data)):
        plt.text(brand_data['Year'].iloc[i], brand_data['Units Sold (million )'].iloc[i],
                 f"{brand_data['Units Sold (million )'].iloc[i]:.2f}",
                 fontsize=10, ha='center', va='bottom')

# เพิ่มชื่อกราฟและแกน
plt.title('แนวโน้มยอดขายโทรศัพท์ตามแบรนด์ (2014-2024)', fontsize=14)
plt.xlabel('ปี', fontsize=12)
plt.ylabel('ยอดขาย (ล้านเครื่อง)', fontsize=12)

# แสดงเส้นกริด
plt.grid(True)

# แสดง Legend
plt.legend(title='แบรนด์')

# แสดงกราฟ
plt.show()
```
